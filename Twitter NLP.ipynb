```python
import pandas as pd
```

```python
df=pd.read_csv("0000000000002747_training_twitter_x_y_train.csv")
df
```

```python
text=df.loc[:,"text"]
text1=list(text)   # text column of the dataframe

y_train=list(df.loc[:,"airline_sentiment"])   # airline setiment column of the dataframe
y_train
```

```python
import nltk
document=[]
for i in range(len(df)):
    text2=nltk.word_tokenize(text1[i])  # converted the text into words
    document.append((text2,y_train[i]))  # tuple with first element as list of words and second element is its airline setiment
document[0]
    
```

```python
from nltk.corpus import stopwords
import string
stops = stopwords.words('english') + list(string.punctuation)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk import pos_tag
```

```python
def clean_review(words):  # function to clean list of words making it free from stop-words and lemmatize it a well
    output_words = []
    for w in words:
        if w.lower() not in stops:
            pos = pos_tag([w])                                 
            clean_word = lemmatizer.lemmatize(w, get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
```

```python
document = [(clean_review(document), category) for document, category in document] 
```

```python
training_documents = document[0:7500]
testing_documents = document[7500:]
```

```python
all_words = []
for doc in document:
    all_words += doc[0]  #collected all words
```

```python
freq = nltk.FreqDist(all_words)                 #will retrurn a freq distribution object
common = freq.most_common(3000)   # took top 3000 words
features = [i[0] for i in common]
features
```

```python
def get_feature_dict(words):   
    current_features = {}         
    words_set = set(words)
    for w in features:
        current_features[w] = w in words_set
    return current_features
```

```python
training_data = [(get_feature_dict(doc), category) for doc, category in training_documents]
testing_data = [(get_feature_dict(doc), category) for doc, category in testing_documents]
```

```python
# classification using NLTK Naive Bayes Classifier
```

```python
from nltk import NaiveBayesClassifier
```

```python
classifier = NaiveBayesClassifier.train(training_data)
```

```python
nltk.classify.accuracy(classifier, testing_data)
```
